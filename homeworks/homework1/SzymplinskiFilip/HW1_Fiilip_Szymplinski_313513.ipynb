## Rozkład ocen dla różnego spożycia alkoholu w weekendy przez uczniów

```python
col_names = ["G1", "G2", "G3"]
table = [(1,"red"),(2,"green"),(3,"blue"),(4,"orange"),(5, "yellow")]
bins = 40
alpha = 1

for col in col_names:
    fig, ax = plt.subplots()
    
    for num, color in table:
        ax.hist(data[data["Walc"]==num][col], bins=bins, alpha=alpha, color=color, label=num)

    ax.set_xlabel(col)
    ax.set_ylabel("Ilość uczniów")

    fig.suptitle(f"Rozkład {col} dla różnych wartośći zmiennej Walc")

    ax.legend()
    plt.show()
```

De facto nie można wyciągnąć żadnych konkretnych wniosków na temat rozkładu ocen końcowych na każdym etapie edukacji w zależności od ilości spożywanego alkoholu w weekend. Występują pewne drobne różnice, ale nie są one na tyle znaczące, aby mów wyciągnąc dalekoidące wnioski.

## Rozkład ocen dla różnego spożycia alkoholu w tygodniu przez uczniów

```python
col_names = ["G1", "G2", "G3"]
table = [(1,"red"),(2,"green"),(3,"blue"),(4,"orange"),(5, "yellow")]

bins = 40
alpha = 1

for col in col_names:
    fig, ax = plt.subplots()
    
    for num, color in table:
        ax.hist(data[data["Dalc"]==num][col], bins=bins, alpha=alpha, color=color, label=num)

    ax.set_xlabel(col)
    ax.set_ylabel("Ilość uczniów")

    fig.suptitle(f"Rozkład {col} dla różnych wartośći zmiennej Dalc")

    ax.legend()
    plt.show()
```

## Rozkład ocen dla różnego liczby niezaliczonych lat nauki

```python
sns.stripplot(data = data, x = "failures", y = "G3", dodge = False, size = 2)
```

```python
col_names = ["G1", "G2", "G3"]
table = [(0,"orange"),(1,"red"),(2,"green"),(3,"blue")]
bins = 40
alpha = 1

for col in col_names:
    fig, ax = plt.subplots()
    
    for num, color in table:
        ax.hist(data[data["failures"]==num][col], bins=bins, alpha=alpha, color=color, label=num)

    ax.set_xlabel(col)
    ax.set_ylabel("Ilość uczniów")

    fig.suptitle(f"Rozkład {col} dla różnych wartośći zmiennej failures")

    ax.legend()
    plt.show()
```

### Mediany ocen dla róznych wartości zmiennej failures

```python
data.loc[:, ["G1", "G2", "G3", "failures"]].groupby(["failures"]).aggregate(np.median)
```

### Średnie ocen dla róznych wartości zmiennej failures

```python
data.loc[:, ["G1", "G2", "G3", "failures"]].groupby(["failures"]).aggregate(np.mean)
```

Spoglądając na wykresy oraz tabele można zauważyć, że oceny są niższe dla większej liczby niezaliczonych lat. Można przyrównać wnioski uzyskane z rozkładów z ujemną koralacją na poziomie około -0.38 zmiennej failures ze zmiennymi G1, G2 oraz G3
